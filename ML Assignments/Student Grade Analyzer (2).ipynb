```python
import pandas as pd
import numpy as np
num_student=1000
num_subject=8
np.random.seed(42)
student=[f"student_{i+1}" for i in range(num_student)]
subject=[f"subject_{j+1}" for j in range(num_subject)]
grades=np.random.randint(40,100,size=(num_student,num_subject))
df=pd.DataFrame(grades,columns=subject)
df.insert(0,"Name",student)
print("===sample of raw data ===")
print(df.head())
```

```python
df["Average"] = df[subject].mean(axis=1).round(2)
def grade_class(avg):
  if avg >=90:
    return "A+"
  elif avg >=80:
      return "A"
  elif avg >=70:
      return "B"
,  elif avg >=60:
      return "C"
  elif avg >=50:
      return "D"
  else:
      return "F"

df["grade"] = df["Average"].apply(grade_class)
df.head()
```

```python
print(">>> Class Statistics <<<")
print("total students",len(df))
print("avg score overall",df[subject].values.mean().round(2))
print("highest avg",df['Average'].max())
print("lowest avg",df['Average'].min())
print("number of A+ student",(df['grade']=="A+").sum(),"\n")
```

```python
print(">> subject wise stats >>")
subject_summary = df[subject].describe().round(2)
print(subject_summary,"\n")
```

```python
top_st =df.nlargest(10,"Average")[["Name","Average","grade"]]
btm_st = df.nsmallest(10,"Average")[["Name","Average","grade"]]

print("top 10 students")
print(top_st,"\n")
print("bottom 10 students")
print(btm_st,"\n")
```

```python
print(">> extra fun analysis >>")

corr = df[subject].corr().round(2)
print("subject Correlation Matrix:")
print(corr,"\n")

highest_mean_sb = df[subject].mean().idxmax()
print(f"subject with highest average score:{highest_mean_sb}")

fail_cnt = (df[subject] < 50).sum()
worst_subject = fail_cnt.idxmax()
print(f"subjects with most fails: {worst_subject}({fail_cnt[worst_subject]}fails)\n")
```

```python
df.to_csv("student_results.csv",index=False)
print("data saved to 'student_result.csv'")
print("\n=== PROGRAM COMMPLETE ===")
```
